# Приведение столбцов к стилю camelCase

```python
def to_camel_case(column_name):
    # Разделяем название по пробелам и подчеркиваниям
    components = column_name.lower().replace('_', ' ').split()
    # Первый компонент оставляем как есть, остальные преобразуем в TitleCase
    return components[0] + ''.join(word.capitalize() for word in components[1:])

# Преобразовать имена столбцов
df.columns = [to_camel_case(col) for col in df.columns]
```

**Динамика выручки сервиса (накопленное Monthly Revenue по месяцу с момента Join Date до момента Last Payment Date)**

```python
# Создать для каждого пользователя временной ряд с месяцами от joinDate до lastPaymentDate

import numpy as np

# Функция для генерации месяцев между двумя датами
def generate_months(start_date, end_date):
    return pd.date_range(start=start_date, end=end_date, freq='MS')  # 'MS' — месяц начала

# Создаем список DataFrame с данными по каждому пользователю и их месяцам
rows = []

for idx, row in df.iterrows():
    months = generate_months(row['joinDate'], row['lastPaymentDate'])
    for month_start in months:
        rows.append({
            'userId': row['userId'],
            'month': month_start,
            'monthlyRevenue': row['monthlyRevenue']
        })

# Создаем DataFrame из списка
monthly_data = pd.DataFrame(rows)
```

```python
# Группируем по месяцу и считаем сумму выручки
monthly_revenue = monthly_data.groupby('month')['monthlyRevenue'].sum().reset_index()
```

```python
# Создать фигуру и две оси
fig, axes = plt.subplots(1, 2, figsize=(16, 6))  # 1 строка, 2 столбца

# График 1: Столбчатая диаграмма
axes[0].bar(
    monthly_revenue['month'].astype(str),
    monthly_revenue['monthlyRevenue'],
    color='skyblue'
)
axes[0].set_xlabel('Месяц')
axes[0].set_ylabel('Общая выручка')
axes[0].set_title('Динамика накопленной выручки сервиса по месяцам')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(True)

# Добавить накопленную выручку, если её нет
if 'cumulativeRevenue' not in monthly_revenue.columns:
    monthly_revenue['cumulativeRevenue'] = monthly_revenue['monthlyRevenue'].cumsum()

# График 2: Линейный график накопленной выручки
axes[1].plot(
    monthly_revenue['month'],
    monthly_revenue['cumulativeRevenue'],
    marker='o',
    color='skyblue'
)
axes[1].set_xlabel('Месяц')
axes[1].set_ylabel('Накопленная выручка')
axes[1].set_title('Кумулятивная динамика выручки')
axes[1].grid(True)

# Подгонка расстояния между графиками
plt.tight_layout()

# Показать графики
plt.show()
```

**Зависимость типа тарифа от группы возраста пользователя**


```python

# Создаем возрастные корзины
bins = [0, 25, 35, 45, 55, 100]
labels = ['18-25', '26-35', '36-45', '46-55', '56+']
df['ageGroup'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

```python
# Вариант А: Таблица с количеством клиентов по типам тарифов и группам

# Таблица абсолютных значений
tariff_counts = pd.crosstab(df['ageGroup'], df['subscriptionType'])
print(tariff_counts)
```

```python
# Вариант Б: Таблица с пропорциями внутри каждой группы (процентное распределение)

# Создаем таблицу пропорций внутри каждой группы (например, по возрастной группе)
tariff_prop = pd.crosstab(df['ageGroup'], df['subscriptionType'], normalize='index')

# Умножаем на 100 для отображения в процентах
tariff_prop_percent = tariff_prop * 100

# Округляем до двух знаков после запятой для читаемости
tariff_prop_percent = tariff_prop_percent.round(2)

print(tariff_prop_percent)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors

tariff_prop_percent = tariff_counts.div(tariff_counts.sum(axis=1), axis=0)

# Цветовая карта для голубых оттенков
blue_map = colors.LinearSegmentedColormap.from_list('blue', ['#ADD8E6', '#00BFFF', '#0000FF'])
blue_colors = ['#ADD8E6', '#00BFFF', '#0000FF']

# Функция для построения столбчатой диаграммы
def plot_bar_chart(data):
    fig, ax = plt.subplots(figsize=(6, 4))  # Уменьшили размер графика
    data.plot(kind='bar', stacked=True, ax=ax, colormap=blue_map)

    # Настраиваем подписи и стили
    plt.xticks(fontsize=8, rotation=0)
    plt.ylabel('Количество клиентов', fontsize=10)  # Уменьшили размер шрифта подписи оси Y
    plt.title('Распределение типов тарифов по группам возраста', fontsize=12)  # Уменьшили размер заголовка
    plt.legend(title='Тип тарифа', fontsize=8)  # Уменьшили размер шрифта легенды

    # Добавляем подписи с количеством клиентов внутри столбцов
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='center', fontsize=6)  # Уменьшили размер шрифта подписей

    plt.tight_layout()
    plt.show()

# Функция для построения круговых диаграмм
def plot_pie_charts(data):
    fig, axes = plt.subplots(1, len(data), figsize=(10, 4))  # Уменьшили размер графика

    for i, (age_group, row) in enumerate(data.iterrows()):
        axes[i].pie(
            row,
            labels=row.index,
            autopct='%.2f%%',
            colors=blue_colors,
            startangle=40
        )
        axes[i].set_title(f'Возрастная группа: {age_group}', fontsize=10)  # Уменьшили размер шрифта заголовков

    # Общий заголовок для всех круговых диаграмм
    fig.suptitle('', fontsize=12)  # Уменьшили размер заголовка

    plt.tight_layout()
    plt.show()

# Построение графиков
plot_bar_chart(tariff_counts)
plot_pie_charts(tariff_prop_percent)
```

```python
# Вариант А: Таблица абсолютных значений (сколько клиентов каждого типа тарифа в каждой стране)

tariff_by_country = pd.crosstab(df['country'], df['subscriptionType'])
# Добавляем столбец с суммарным количеством клиентов по каждой стране
tariff_by_country['total'] = tariff_by_country.sum(axis=1)

# Сортируем по этому столбцу (по убыванию)
tariff_by_country_sorted = tariff_by_country.sort_values(by='total', ascending=False)

# Удаляем временный столбец, если он больше не нужен
tariff_by_country_sorted = tariff_by_country_sorted.drop(columns='total')
print(tariff_by_country)
```

```python
# Вариант Б: Таблица с пропорциями внутри каждой страны (процентное распределение)

# Таблица пропорций внутри каждой страны
tariff_prop_country = pd.crosstab(df['country'], df['subscriptionType'], normalize='index')

# Отформатируем значения как проценты с 1 или 2 знаками
tariff_prop_country_percent = tariff_prop_country * 100
tariff_prop_country_formatted = tariff_prop_country_percent.round(2)

print(tariff_prop_country_formatted)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Создаем таблицу подсчета общего количества клиентов по странам
country_total = df['country'].value_counts()

# Сортируем страны по убыванию общего количества клиентов
sorted_countries = country_total.sort_values(ascending=False).index

# Создаем таблицу пропорций внутри каждой страны
tariff_by_country = pd.crosstab(df['country'], df['subscriptionType'])

# Перестраиваем таблицу в порядке сортировки стран
tariff_by_country_sorted = tariff_by_country.loc[sorted_countries]

# Указываем цвета для графика
blue_colors = ['#ADD8E6', '#00BFFF', '#0000FF']  # Светло-голубой, яркий голубой, синий

# Строим стековую диаграмму
ax = tariff_by_country_sorted.plot(kind='bar', stacked=True, figsize=(12, 8), color=blue_colors)
plt.ylabel('Количество клиентов')
plt.title('Распределение типов тарифов по странам')
plt.legend(title='Тип тарифа')

# Добавляем подписи с количеством клиентов внутри сегментов
for container in ax.containers:
    ax.bar_label(container, label_type='center', fontsize=8)

# Добавляем сумму по каждому столбцу (общее число клиентов в стране)
for idx, country in enumerate(tariff_by_country_sorted.index):
    total = country_total[country]
    # Размещаем подпись чуть выше верхней границы столбца
    ax.text(idx, total + max(country_total) * 0.01, str(total), ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```
